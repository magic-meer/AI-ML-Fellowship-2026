**📝 Reflection:**

Which tokenizer produced more tokens?

Why?

**Answer:** BERT tokenizer typically produces more tokens than GPT2 for the same text. This is because:

1. **Subword tokenization**: BERT uses WordPiece tokenization which breaks words into smaller subword units
2. **Special tokens**: BERT adds [CLS], [SEP], [PAD] tokens
3. **Vocabulary size**: GPT2 has a larger vocabulary (50k) vs BERT (30k), allowing GPT2 to handle more words as single tokens
4. **Tokenization rules**: They use different tokenization algorithms with different splitting strategies

This affects cost (API pricing is often per-token) and context window usage.

## Section 3 — RAG Fundamentals (Day 4)
### What is Retrieval Augmented Generation (RAG)?
**RAG** = Retrieval + Generation. It's a technique where an AI model retrieves relevant information from a knowledge base before generating an answer. This allows LLMs to answer questions about data they weren't trained on.

### Why LLMs forget information
- LLMs have a fixed knowledge cutoff - they can't know about events after training
- They have limited context windows (can't read your entire document)
- They can hallucinate (make up facts) when uncertain
- RAG solves this by providing real-time, accurate context

### Embeddings concept
Embeddings are numerical representations of text that capture semantic meaning. Similar concepts have similar vectors. Example: "cat" and "dog" have similar embeddings (both are pets), while "car" is far apart.

### Vector search
Vector search finds the most similar documents to a query by comparing their embedding vectors. We use cosine similarity or dot product to measure closeness. The most similar (highest score) documents are retrieved as context for the LLM.

```python
# Step 1: Create Documents
docs = [
    "AI stands for Artificial Intelligence.",
    "Machine learning is a subset of AI.",
    "GANs generate synthetic data."
]
```

```python
# Step 2: Create Embeddings (FREE)
from sentence_transformers import SentenceTransformer

embed_model = SentenceTransformer('all-MiniLM-L6-v2')

doc_embeddings = embed_model.encode(docs)
```

```python
# Step 3: Search Relevant Text
import numpy as np

query = "What is AI?"
query_embedding = embed_model.encode([query])

scores = np.dot(doc_embeddings, query_embedding.T)

best_doc = docs[np.argmax(scores)]
print(best_doc)
```

```python
# Extended RAG with 10 documents

# Step 1: Create 10 documents
docs_extended = [
    "Artificial Intelligence (AI) refers to the simulation of human intelligence in machines programmed to think and learn like humans.",
    "Machine Learning (ML) is a subset of AI that enables systems to automatically learn and improve from experience without being explicitly programmed.",
    "Deep Learning is a subset of ML that uses neural networks with multiple layers to solve complex problems.",
    "Natural Language Processing (NLP) is a branch of AI that helps computers understand, interpret, and manipulate human language.",
    "Computer Vision is an AI field that enables computers to derive meaningful information from digital images and videos.",
    "Reinforcement Learning is a type of ML where an agent learns to make decisions by taking actions in an environment to maximize a reward.",
    "Generative AI refers to AI algorithms that can create new content including audio, code, images, text, simulations, and videos.",
    "Large Language Models (LLMs) are AI models trained on vast amounts of text data to understand and generate human-like language.",
    "Transfer Learning is a ML technique where a model developed for one task is reused as the starting point for a model on a second task.",
    "Transformer models are neural network architectures that use self-attention mechanisms, enabling them to process sequential data efficiently."
]

# Step 2: Create embeddings for all documents
doc_embeddings_extended = embed_model.encode(docs_extended)

print(f"Created embeddings for {len(docs_extended)} documents")
print(f"Embedding shape: {doc_embeddings_extended.shape}")
```

```python
# Step 3: Custom question answering
def rag_query(query_text, documents, embeddings, embed_model, model):
    """Simple RAG function to answer questions from documents"""
    # Encode query
    query_embedding = embed_model.encode([query_text])

    # Calculate similarity scores
    scores = np.dot(embeddings, query_embedding.T)

    # Get top 2 most relevant documents
    # Flatten scores before argsort to ensure top_indices is a 1D array of integers
    top_indices = np.argsort(scores.flatten())[-2:][::-1]

    # Combine relevant context
    context = "\n".join([documents[i] for i in top_indices])

    # Generate answer using LLM
    prompt = f"Based only on the following context, answer the question.\n\nContext:\n{context}\n\nQuestion: {query_text}\n\nAnswer:"
    response = model.generate_content(prompt)

    return response.text, context

# Test with custom questions
questions = [
    "What is the difference between AI and Machine Learning?",
    "How do Transformer models work?",
    "What is Generative AI used for?"
]

for q in questions:
    answer, context = rag_query(q, docs_extended, doc_embeddings_extended, embed_model, model)
    print(f"Question: {q}")
    print(f"Answer: {answer}\n")
```

### Simple GAN (PyTorch)


```python
import torch
import torch.nn as nn
```

#### Generator


```python
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(100,128),
            nn.ReLU(),
            nn.Linear(128,784),
            nn.Tanh()
        )

    def forward(self,x):
        return self.model(x)
```

#### Discriminator



```python
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(784,128),
            nn.ReLU(),
            nn.Linear(128,1),
            nn.Sigmoid()
        )

    def forward(self,x):
        return self.model(x)
```

**📝 Task:**

Explain generator vs discriminator roles.

### Generator vs Discriminator Roles

**Generator:**
- Creates fake data (images, text, etc.) from random noise
- Learns to produce outputs that look real
- Goal: Fool the discriminator

**Discriminator:**
- Acts as a judge, classifying inputs as real or fake
- Learns to distinguish between real and generated data
- Goal: Correctly identify fake samples

**Training Process:**
1. Generator creates fake samples
2. Discriminator evaluates them alongside real samples
3. Both networks improve through competition
4. Eventually, generator produces highly realistic outputs

### Data Augmentation using GANs

**Concept Task:**

**Reflection questions:**

- Why use GAN augmentation?

- When is synthetic data risky?

**Answer:**

**Why use GAN augmentation?**
- Increases dataset diversity without collecting new real data
- Helps balance imbalanced datasets (e.g., rare disease detection)
- Can improve model generalization and robustness
- Useful when real data is expensive or difficult to collect

**When is synthetic data risky?**
- When GAN mode collapse occurs (produces limited variety)
- If training data has biases, GAN will replicate them
- When domain shift exists between synthetic and real data
- In high-stakes domains (medical, legal) where errors are costly
- When regulatory compliance requires real-world validation

```python
# GAN Data Augmentation Example
# In practice, you would train a GAN on real data to generate synthetic samples

# For demonstration, let's show how to generate multiple synthetic samples
gen = Generator()

# Generate 5 synthetic images
synthetic_samples = []
for i in range(5):
    noise = torch.randn(1, 100)
    fake = gen(noise)
    synthetic_samples.append(fake)
    print(f"Generated synthetic sample {i+1}: shape {fake.shape}")

print("\nThese synthetic samples can be added to training datasets to:")
print("- Increase dataset diversity")
print("- Balance imbalanced classes")
print("- Improve model generalization")

print("\n⚠️ Risks of synthetic data:")
print("- If GAN mode collapses, samples may be too similar")
print("- May introduce biases present in training data")
print("- Quality depends on GAN training quality")
```

🎯 **Build:** Text Analysis Tool

```python
# Text Analysis Tool - Complete Implementation

from transformers import AutoTokenizer

# 1. Token Counter Function
def count_tokens(text, model_name="gpt2"):
    """Count the number of tokens in text using a specific tokenizer"""
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokens = tokenizer.encode(text)
    return len(tokens), tokens

# 2. Summarizer Function (using Gemini)
def summarize_text(text, model):
    """Summarize text using Gemini"""
    prompt = f"Provide a concise summary of the following text:\n\n{text}"
    response = model.generate_content(prompt)
    return response.text

# 3. Combined Tool
def analyze_text(text, model, tokenizer_model="gpt2"):
    """Complete text analysis: token count + summary"""
    token_count, tokens = count_tokens(text, tokenizer_model)
    summary = summarize_text(text, model)

    return {
        "token_count": token_count,
        "summary": summary
    }

# Test the Text Analysis Tool
sample_text = """
Machine learning is a subset of artificial intelligence that focuses on
teaching computers to learn from data and improve their performance
over time without being explicitly programmed. It has become one of
the most important technologies of the 21st century, powering everything
from recommendation systems to autonomous vehicles. Deep learning,
a further subset of machine learning, uses neural networks with many
layers to achieve remarkable results in image recognition, natural
language processing, and speech recognition.
"""

result = analyze_text(sample_text, model)
print("=== Text Analysis Results ===")
print(f"Token Count: {result['token_count']}")
print(f"Summary: {result['summary']}")
```
